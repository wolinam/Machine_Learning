```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
# Series를 통한 line plot 그리기
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10)) #랜덤값 10개. index는 1부터 100까지 10단위
# rand : 0~1사이의 난수
# randn : 가우시안 표준 정규 분포에서의 난수
# randint : 정수 난수
s
```

```python
s.plot()
```

```python
# DataFrame을 통한 line plot
# randn(m,n) : mxn 크기의 난수 행렬
# cumsum : 누적합
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0),
                  columns=["A", "B", "C", "D"],
                  index=np.arange(0, 100, 10))
df
```

### plot

```python
df.plot()
# 하나의 열에 대해서만 보고 싶다면
# df['B'].plot()
```

### bar

```python
df.plot(kind='bar')
df.plot(kind='barh') #horizon
```

```python
df2 = pd.DataFrame(np.random.rand(6, 4), 
                   index=["one", "two", "three", "four", "five", "six"],
                   columns=pd.Index(["A", "B", "C", "D"], name="Genus"))
df2
```

```python
df2.plot(kind='bar')
```

```python
df2.plot(kind='barh', stacked=True)
```

### histogram

```python
# histogram은 index가 필요없음

fig = plt.figure()
s3 = fig.add_subplot(2, 2, 1)
s4 = fig.add_subplot(2, 2, 2)
s5 = fig.add_subplot(2, 2, 3)
s3.hist(pd.Series(np.random.normal(0, 1, size=200))) #0부터 200까지 1씩 증가하는 단위
s4.hist(pd.Series(np.random.normal(0, 1, size=200)),bins=50)
s5.hist(pd.Series(np.random.normal(0, 1, size=200)),bins=100, normed=True)
# normed 속성을 True로 설정하면, 각 bin에 속하는 개수를 전체 개수로 나눈 비율, 즉 정규화 한 값을 bar의 높이로 사용하게 된다.
```

```python
x1 = np.random.normal(1, 1, size=(100, 1))
x1
x2 = np.random.normal(-2, 4, size=(100, 1))
X = np.concatenate((x1, x2), axis=1) #배열 2개를 열의 방향으로 하나로 합침
X
```

```python
df3 = pd.DataFrame(X, columns=["x1", "x2"])
df3
```

```python
fig0 = plt.figure()
plt.scatter(df3['x1'], df3['x2']) #x1이 x축, x2가 y축
```
